# Weak acid speciation versus pH

In class we derived the equations for $\alpha$ versus pH

$\alpha_{HA}=\dfrac{[\mathrm{H}^+]}{[\mathrm{H}^+]+K_a}$

$\alpha_{A}=\dfrac{K_a}{[\mathrm{H}^+]+K_a}$

The script below will request a pKa value from the user and then plot the corresponding alpha diagram.

```octave
pKa = input("Enter pKa1 value: "); Ka=10^-pKa;
%pKa=4; Ka=10^-pKa;
pH=2:0.2:12; H=10.^-pH;
alphaHL=H./(Ka+H);
alphaL=Ka./(Ka+H);
h1=plot(pH,alphaHL,'r-','linewidth',2); hold on
h2=plot(pH,alphaL,'b-',"linewidth", 2);
set(gca,'linewidth',2,'fontsize',12)
xlabel('pH')
ylabel('\alpha')
plot(pH,alphaHL,pH,alphaL)
legend('HA','A^-')
```
