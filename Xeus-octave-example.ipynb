```octave
graphics_toolkit("plotly")

a = [1:0.1:13];
theta = linspace (0,2*pi,1000);
rho = sin (7*theta);

subplot(2,2,1);
sombrero()
grid on;
box on;

subplot(2,2,2);
plot(theta, rho);
grid on;
box on;

subplot(2,2,3);
polar(a,a)
grid on;
box on;

subplot(2,2,4);
polar(theta, rho);
grid on;
box on;
```
