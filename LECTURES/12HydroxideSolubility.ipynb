# Hydroxide Solubility  
### Iron and aluminum hydroxides solids  
$\mathrm{Fe(OH)_3 (s)}$ &nbsp;&nbsp;&nbsp;&nbsp; Ksp = 10<sup>-38.8</sup>  
For Fe<sup>3+</sup>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Ba<sub>1</sub> = 10<sup>-2.19</sup>  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Ba<sub>2</sub> = 10<sup>-5.7</sup>  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Ba<sub>4</sub> = 10<sup>-21.6</sup>  

$\mathrm{Ksp = [F^{3+}][OH^-]}^3$  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $=\mathrm{\frac{[Fe^{3+}]Kw^3}{[H^+]^3}}$  
$\mathrm{[Fe^{3+}] = \frac{Ksp[H^+]^3}{Kw^3}}$  
$\mathrm{log[Fe^{3+}] = log(\frac{Ksp}{Kw^3}) + 3log[H^+]}$  
$\mathrm{log[Fe^{3+}] = log(\frac{Ksp}{Kw^3}) - 3pH}$  
y = b + mx

Ba_1&nbsp;&nbsp;&nbsp;&nbsp; $\mathrm{Fe^{3+} + H_2O = FeOH^{2+} + H^+}$  
$\mathrm{Ba_1 = \frac{[FeOH^{2+}][H^+]}{[Fe^{3+}]}}$  
remember: $\mathrm{[Fe^{3+}] = \frac{Ksp*[H^+]^3}{Kw^3}}$  
sub in $\mathrm{Ba_1 = \frac{[FeOH^{2+}][H^+]Kw^3}{Ksp[H^+]^3}}$  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $=\mathrm{\frac{[FeOH^{2+}]Kw^3}{Ksp[H^+]^2}}$  
$\mathrm{[FeOH^{2+}] = \frac{Ba_1 - Ksp[H^+]^2}{Kw^3}}$  
$\mathrm{log[FeOH^{2+}] = log(\frac{Ba_1Ksp}{Kw^3}) + 2log[H^+]}$  
$\mathrm{log[FeOH^{2+}] = log(\frac{Ba_1Ksp}{Kw^3}) + 2pH}$

So we figured out the equations for Fe<sup>3+</sup> and FeOH<sup>2+</sup> in equillibrium with Fe(OH)<sub>3(s)</sub>. Next: Fe(OH)<sub>2</sub><sup>+</sup> in equillibrium with Fe(OH)<sub>3(s)</sub>.  
We discovered that: $\mathrm{[Fe^{3+}] = \frac{Ksp*[H^+]^3}{Kw^3}}$

$\mathrm{Fe^{3+} + 2H_2O = 2H^+ + Fe(OH)_2^+}$  
$\mathrm{Ba_2 = \frac{[H^+]^2[Fe(OH)_2^+]}{[Fe^{3+}]}}$   
$\mathrm{[Fe(OH)_2^+] = \frac{Ba_2*[Fe^{3+}]}{[H^+]^2}}$  
sub in $\mathrm{Fe^{3+}}$ from last time $\mathrm{= \frac{Ba_2Ksp[H^+]^3}{Kw^3[H^+]^2}}$  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $\mathrm{= \frac{Ba_2Ksp[H^+]}{Kw^3}}$  
$\mathrm{log[Fe(OH)_2^+] = log(\frac{Ba_2Ksp}{Kw^3}) + log[H^+]}$  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$\mathrm{= log(\frac{Ba_2Ksp}{Kw^3}) - pH}$

$\mathrm{y = log[Fe(OH)_2^+]}$  
$\mathrm{b = log(\frac{Ba_2Ksp}{Kw^3})}$  
$\mathrm{m = -1}$  
$\mathrm{x = pH}$

$\mathrm{Fe(OH)_4}^-$  
$\mathrm{Ba_4 Fe^{3+} + 4H_2O = Fe(OH)_4^- + 4H^+}$  
$\mathrm{Ba_4 = \frac{[Fe(OH)_4^-][H^+]^4}{[Fe^{3+}]}}$ &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; sub in for [Fe<sup>3+</sup>]  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $\mathrm{= \frac{[Fe(OH)_4^-]Kw^3[H^+]^4}{Ksp[H^+]^3}}$  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$\mathrm{= \frac{[Fe(OH)_4^-]Kw^3[H^+]}{Ksp}}$  
$\mathrm{[Fe(OH)_4^-] = \frac{Ba_4Ksp}{Kw^3[H^+]}}$  

aside:  
$\mathrm{Ba_4 = \frac{[Fe(OH)_4^-][H^+]^4}{[Fe^{3+}]}}$  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$=\mathrm{[Fe(OH)_4^-][H^+]^4 * \frac{1}{[Fe^{3+}]}}$  
$\mathrm{Fe^{7+} = \frac{Ksp[H^+]^3}{Kw^3}}$  
$\mathrm{\frac{1}{Fe^{3+}} = \frac{Kw^3}{Ksp [H^+]^3}}$

$\mathrm{log[Fe(OH)_4^- = log(\mathrm{Ba_4 * Ksp}{Kw^3})}$  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$\mathrm{-log[H^+]}$  
the slope is now +1

```octave
%plot -s 600,500 -f 'svg'
pH=0:0.1:12;
Ksp=10^-38.8;
B1=10^-2.19;
B2=10^-5.7;
B4=10^-21.6;
Kw=1e-14;
logFe=log10(Ksp./Kw^3)-3*pH; Fe=10.^logFe;
logFeOH=log10(B1*Ksp/Kw^3)-2*pH; FeOH=10.^logFeOH;
logFeOH2=log10(B2*Ksp/Kw^3)-pH; FeOH2=10.^logFeOH2;
logFeOH4=log10(B4*Ksp./Kw^3)+pH; FeOH4=10.^logFeOH4;
FeT=Fe+FeOH+FeOH2+FeOH4; logFeT=log10(FeT);

plot(pH,logFe,'linewidth',2,pH,logFeOH,'linewidth',2,pH,logFeOH2,'linewidth',2,...
pH,logFeOH4,'linewidth',2,pH,logFeT,'linewidth',2)
hold on
plot([-log10(B1) -log10(B1)],[-40 10],'k--','linewidth',2)
set(gca,'linewidth',2,'fontsize',12);
xlabel('pH','fontsize',12)
ylabel('log[species]','fontsize',12)
legend('Fe^{3+}','FeOH^{2+}','Fe(OH)2^+','Fe(OH)4^-','Fe_T','location','eastoutside');
```

<u>Saturation Index</u>  
$\mathrm{Ksp = [A^+]^a_{eq}[B-]^b_{eq}}$  
IAP is the ion activity product.  
$\mathrm{= [A^+]^a_{actual}[B-]^b_{actual}}$  
SI is the saturation index.  
$=\mathrm{\frac{IAP}{Ksp}}$

SI = 0 &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; equillibrium  
&nbsp;&nbsp;&nbsp; > 0 &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; supersaturated  
&nbsp;&nbsp;&nbsp; < 0 &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; undersaturated
