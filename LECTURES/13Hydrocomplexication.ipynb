```octave
%plot -s 600,500 -f 'svg'
logB1=4.1; logB2=7.7; logB3=10.3; logB4=12; Kw=1e-14; logKw=log10(Kw);
logBa1=logB1+logKw; logBa2=logB2+2*logKw; logBa3=logB3+3*logKw; logBa4=logB4+4*logKw;
Ba1=10^logBa1; Ba2=10^logBa2; Ba3=10^logBa3; Ba4=10^logBa4;

pH=6:0.1:14; H=10.^(-pH); OH=Kw./H;
denominator=H.^4+Ba1*H.^3+Ba2*H.^2+Ba3*H+Ba4;
alphaCd=(H.^4)./denominator;
alphaCdOH=(H.^3*Ba1)./denominator;
alphaCdOH2=(H.^2*Ba2)./denominator;
alphaCdOH3=(H*Ba3)./denominator;
alphaCdOH4=(Ba4)./denominator;

plot(pH,alphaCd,'b','linewidth',2)
hold on
plot(pH,alphaCdOH,'g','linewidth',2)
plot(pH,alphaCdOH2,'m','linewidth',2)
plot(pH,alphaCdOH3,'k','linewidth',2)
plot(pH,alphaCdOH4,'c','linewidth',2)
xlabel('pH'); ylabel('\alpha');
legend('Cd^{2+}','CdOH^+','Cd(OH)2','Cd(OH)3^-','Cd(OH)4^{2-}','location','west')
set(gca,'linewidth',2)
title('Distribution diagram, Cd hydroxo complexation')

% add stepwise values (pH=pKa but we need to convert because we were given overall association constants)

plot([9.9 9.9],[0 1],'k:','linewidth',2)
plot([10.4 10.4],[0 1],'k:','linewidth',2)
plot([11.4 11.4],[0 1],'k:','linewidth',2)
plot([12.3 12.3],[0 1],'k:','linewidth',2)
```

$\mathrm{Ba_1 = \frac{((dOH][H]}{[(d]}}$ &nbsp;&nbsp;&nbsp;&nbsp; $\mathrm{Ba_2 = \frac{((dO(H)_2][H]^2}{[(d]}}$
