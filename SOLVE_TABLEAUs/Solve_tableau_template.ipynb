# Instructions to run tableau

1. change the TABLEAU entries below.  as demonstrated in lecture.  The % symbols are comments but useful for you to keep track of what species you are inputting for each row.  Do not change any thing else.  If you make a mistake simply close your browser and reclick on the link to start again from the original template.
2. when you are happy with the TABLEAU go to the run menu at the top and select "Run All Cells"
3. the outputs will display as species1, species2, ... in the final cell where the order of species is the exact same as you input them in the TABLEAU

```octave
% ------ CHANGE ONLY BETEWEEN THESE LINES %-----------------------------------------------------------
TABLEAU=[...
%H      A        logK  species
1       0        0     %H
0       1        0     %A
-1      0        -14   %OH
1       1        4.75  %HA
0.01    0.01     0     %TOTALs
];
%------ CHANGE ONLY BETEWEEN THESE LINES %-----------------------------------------------------------
```

# The tableau is solved and results output using the cells below

```octave
[N,M]=size(TABLEAU);

ASOLUTION=TABLEAU(1:N-1,1:M-1);
KSOLUTION=TABLEAU(1:N-1,M);
T=TABLEAU(N,1:M-1)'; T(T==0)=1e-30;

format long e

X=T./10; [xans,masserr,J,C] = nl_massbalancerrnosolid_NR(X,ASOLUTION,KSOLUTION,T);
```

```octave
% display with lots of digits

for i=1:length(C)
    istr=num2str(i); 
    concstr=num2str(C(i),16);
    text=['species',istr,'=',concstr,' M'];
    disp(text)
end
```
